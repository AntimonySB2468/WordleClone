```python
__author__ = "A. Anand"

# Functions made by Mr. Banjevic:

import random

def remaining_letters(text):
    """Return text coloured black, with gray background. Remaining letters."""

    return "\033[0;30;47m"+text+"\033[0m"


def wrong_letter(text):
    """Return text coloured white with black background. Not in word."""

    return "\033[5;37;40m"+text+"\033[0m"


def correct_letter(text):
    """Return text coloured magenta with green background. Correct Position."""

    return "\033[0;35;42m"+text+"\033[0m"

def wrong_position(text):
    """Return text coloured black with yellow background. In word, wrong position."""

    return "\033[0;30;43m"+text+"\033[0m"


def get_word(word_list):
    """Return a word randomly chosen from word_list."""

    return random.choice(word_list)


# Functions made by me:

def game_instructions():
  '''Print wordle instructions.'''

  print("How to play:")
  print("• Guess a 5 letter word in 6 tries. \n• The colour of the word will show how close your guess was.\n")
  print("{} E A R Y".format(correct_letter("W")))
  print("W is in the word and in the correct spot.\n")
  print("P {} L L S".format(wrong_position("I")))
  print("I is in the word but in the wrong spot.\n")
  print("V A G {} E".format(wrong_letter("U")))
  print("U is not in the word in any spot.\n")


def clear():
  '''Clear the whole screen.'''

  from google.colab import output
  return output.clear()


def user_input(msg, quit, help):
  '''
  Get input from the user and return either:
  > the quit string 'q' if requesting to quit
  > the help string 'h' if requesting instructions
  > False if it's lowercase version doesn't fit the conditions of not being 5
    letters long, or not in the list of words
  > The user's input
  '''

  user_guess = input(msg)

  if user_guess.lower() == quit:
    return 'q'
  elif user_guess.lower() == help:
    return 'h'
  elif len(user_guess) != 5 or user_guess.lower() not in WORDS:
    return False
  else:
    return user_guess


def progress(guess_list, alpha_list, attempts, games):
  '''
  Print the game number, the number of guesses remaining, the list of
  guesses, and the list of letters along with their colours.
  '''

  clear()
  print("Game:", games)

  print("Guesses remaining: {}".format(attempts))
  for guess in guess_list:
    print(guess)

  print("\n\nLetters:")
  for j in range(len(alpha_list)):
    if j % (len(alpha_list)//2) == 0:
      print("\n", end ='')
    print("{} ".format(alpha_list[j]), end ='')
  print()


def print_win(score, score_list):
  '''
  Print the distribuition of each score.
  '''

  print("{}:".format(score),end='')
  for win in score_list:
    if win == score:
      print("■", end = '')
  print()


def force_agree(msg, yes, no):
    '''Force the variable to confine to reply either yes or no.'''

    agree = input(msg)

    while agree.lower() not in [yes, no]:
        print("That's an invalid answer.")
        agree = input(msg)
    return agree


# Game:

def wordle():

  print("Welcome to Aadhya's Wordle Clone\n")

  game_instructions() # Prints the game's instructions.

  agree = force_agree("Do you want to play? Enter y/n \n", 'y', 'n')

  games = 0

  scores = []

  while agree.lower() == "y":
    word = get_word(WORDS).upper()

    guess_list = ["_ _ _ _ _", "_ _ _ _ _", "_ _ _ _ _", "_ _ _ _ _", "_ _ _ _ _", "_ _ _ _ _"] # This is the list of guesses.

    # The alphabet list:
    letters = []
    for i in range(65, 91):
      letters.append(remaining_letters(chr(i)))

    attempts = 6

    letters_correct = [False, False, False, False, False] # This list keeps in track all of the guessed letters in the game.

    play = True

    while attempts > 0 and play:
      progress(guess_list, letters, attempts, games) # Prints the game.

      user_guess = user_input("\n\nEnter a five letter word. \nEnter q to quit. \nEnter h to view instructions \n", "q", "h")

      colour_guess = '' # This stores each letter of the guessed word including the colour.

      if not(user_guess): # If the user doesn't give proper input.
        print("Invalid input.")

        agree = force_agree("Do you wish to continue? Enter y/n \n", 'y', 'n')

      elif user_guess == 'q': # If the user wishes to quit.
        play = False

      elif user_guess == 'h': # If the user wants to view instructions during the game.
        clear()

        game_instructions() # Prints out the game instructions should the user need help.

        agree = force_agree("Do you wish to continue? Enter y/n \n", 'y', 'n')

      else: # The following code colours each letter in the guess according to the rules of wordle.

        guess = user_guess.upper() # This makes the following code easier to understand.

        for i in range(5): # Since the guess is already only allowed if they are 5 letters long, I can simply put in this for loop with range(5) for each letter.

          # In word and right position.
          if guess[i] == word[i]:

            letters_correct[i] = True
            colour_guess += correct_letter(guess[i]) + ' '

            # Recolours the remaining letters in the alphabet list to green.
            if remaining_letters(guess[i]) in letters:
              letters[letters.index(remaining_letters(guess[i]))] = correct_letter(guess[i])

            # Recolours the yellow letters in the alphabet list to green.
            elif wrong_position(guess[i]) in letters:
              letters[letters.index(wrong_position(guess[i]))] = correct_letter(guess[i])


          else:

            # In word but wrong position:
            if guess[i] in word:
              colour_guess += wrong_position(guess[i]) + ' '

              # Recolours the remaining letters in the alphabet list to yellow.
              if remaining_letters(guess[i]) in letters:
                letters[letters.index(remaining_letters(guess[i]))] = wrong_position(guess[i])


            # Not in word:
            else:
              colour_guess += wrong_letter(guess[i]) + ' '

              # Recolours the remaining letters in the alphabet list to black.
              if remaining_letters(guess[i]) in letters:
                letters[letters.index(remaining_letters(guess[i]))] = wrong_letter(guess[i])

        print()

        guess_list[6-attempts] = colour_guess # Adds the coloured letters of the guessed word to the guess list.

        if letters_correct == [True, True, True, True, True]: # Ends game if all letters are correct.
            play = False

        else: # Deducts number of attempts remaining.
            attempts -= 1

    games += 1

    if False in letters_correct: # If the user doesn't manage to guess the word, this prints.
      print("\nThe word was {}".format(word))

      scores.append(0)

    else:
      print("Congratulations! You correctly guessed the word!")

      print("The word was {}".format(word))

      scores.append(6-attempts)

    # The user has the choice to play again if they wish.
    agree = force_agree("\nWould you like to play again? Enter y/n \n", 'y', 'n')


  clear()

  print("Thanks for playing!\n")

  print("Statistics:")

  print("\nNumber of games played:", games)

  print("\nWin distribuition:")
  for i in range(7):
    print_win(i, scores)


if __name__ == '__main__':
  wordle()
```
